##Data Pre-processing

###Loading the data

```python
import pandas as pd
df = pd.read_csv('diabetes-dataset.csv')
```

```python
df.head()
```

```python
# To check for missing values

df.isnull().sum()
```

It means there are no null values in any of the columns.

```python
# Now, I am going to find number of diabetes and non-diabetes persons
print(df.Outcome.value_counts())
```

###Feature selection

Here, we will take the features that impact the outcome column.

```python
# select feature colulmn
feature_columns = df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
      'BMI', 'DiabetesPedigreeFunction', 'Age']]
feature_columns.head()
# Outcome colummn
outcome_column = df['Outcome']
outcome_column.head()
```

###Splitting Train and Test data

From the collected data, we are splitting the data into two parts for train and test in the ratio of 80-20.

```python
# import package
from sklearn.model_selection import train_test_split
# split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(feature_columns, outcome_column, test_size=0.2, random_state=5)
```

###Training Data – Combining X_train and y_train

```python
training_data = X_train
training_data['outcome'] = y_train.values
```

```python
training_data
```

###Testing Data – Combining X_test and y_test

```python
test_data = X_test
test_data['outcome'] = y_test.values
```

```python
test_data
```

###Converting data to **CSV**

Now, I convert this training data and testing data frames into CSV files. Because when I will creating a pipeline, I have to import this data into google cloud storage bucket. For that, it should be in some supported formats like CSV.

```python
training_data.to_csv("Training_data.csv")
test_data.to_csv("Test_data.csv")
```
